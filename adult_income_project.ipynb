```python
X = df.drop('income', axis=1)
y = df['income']
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler,RobustScaler, OneHotEncoder

from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=42)
```

```python
print("X_train: ", X_train.shape)
print("y_train: ", y_train.shape)
print("X_test:  ", X_test.shape)
print("y_test:  ", y_test.shape)
```

```python
cat = X_train.select_dtypes("object").columns
cat
```

```python
encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
transformer = make_column_transformer((encoder, cat), remainder=StandardScaler())
```

<h2 style="background-color: #FAF9F6; font-family: newtimroman; font-size: 250%; color: darkblue; border: 12px solid #A9A9A9; border-radius: 20px; padding: 15px; text-align: center;">
    <b>Logistic Regression</b>
</h2>

```python
from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import Pipeline ,make_pipeline
from sklearn.model_selection import GridSearchCV


from sklearn.metrics import make_scorer, f1_score
```

```python
penalty = ["l1", "l2"]
C = np.logspace(-1, 5, 10)
class_weight = ["balanced", None]
solver = ["lbfgs", "liblinear", "sag", "saga"]

param_grid = {
    "log_model__penalty": penalty,
    "log_model__C": C, 
    "log_model__class_weight": class_weight,
    "log_model__solver": solver
}

scorer = make_scorer(f1_score, average='macro')

steps = [("column_trans", transformer),
         ("scaler", StandardScaler(with_mean=False)),
         ("log_model", LogisticRegression(max_iter=10000, random_state=42))]
```

```python
pipe_model = Pipeline(steps)
```

```python
log_grid_model = GridSearchCV(estimator=pipe_model,
                              param_grid=param_grid,
                              scoring=scorer,
                              return_train_score=True,
                              cv=10,
                              n_jobs=-1)
```

```python
log_grid_model.fit(X_train, y_train)
```

```python
log_grid_model.best_params_
```

```python
log_grid_model.best_score_
```

<div class="alert alert-warning alert-info" style="background-color: #FAF9F6; border: 5px solid #A9A9A9; border-radius: 10px;">
    <h3 style="font-weight: bold; color: darkblue;">Logistic Model Performance</h3>
</div>

```python
from sklearn.metrics import confusion_matrix, classification_report

def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)

    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
    print("---------------------------------------------")
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))    
```

```python
eval_metric(log_grid_model, X_train, y_train, X_test, y_test)
```
